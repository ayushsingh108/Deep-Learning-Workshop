# Hidden Layer using Sigmoid Activation

## 🎯 Objective

Compute the output of a hidden neuron.

---

## 📚 Theory

A hidden neuron calculates a weighted sum and passes it through the Sigmoid function.

---

## 🧮 Formula

z = Σ(wx)+b

Output = Sigmoid(z)

---

## 💻 Implementation

---

## 📊 Output

---

## 📖 Conclusion

The hidden neuron transforms linear inputs into nonlinear outputs.

```python
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))
x1 = 0.5
x2 = 0.8
w1 = 0.4
w2 = 0.7
b1 = 0.2
z = x1*w1 + x2*w2+b1
hidden_output = sigmoid(z)
print("Weight Sum =", round(z,3))
print("Hidden Layer Output =", round(hidden_output,3))
```
