# Sigmoid Learning using Backpropagation

## 🎯 Objective

Update the weight of a Sigmoid neuron using gradient descent.

---

## 📚 Theory

Backpropagation computes gradients and updates weights to reduce prediction error.

---

## 🧮 Formula

Sigmoid Derivative

σ(x)(1−σ(x))

Weight Update

W = W + η × Gradient

---

## 💻 Implementation

---

## 📊 Output

---

## 📖 Conclusion

The neuron gradually learns by minimizing error.

```python
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_derivative(y):
  return y*(1-y)

x=1
target = 1
weight = 0.5
learning_rate = 0.5

for epoch in range(10):
  z=x*weight
  output = sigmoid(z)
  error = target - output
  gradient =  error * sigmoid_derivative(output)
  weight = weight + learning_rate * gradient * x
  print(f"Epoch {epoch+1}")
  print("Output =", round(output,3))
  print("Error =", round(error,3))
  print("Weight =", round(weight,3))
  print("------------------------")
```
