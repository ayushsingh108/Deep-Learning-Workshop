# Question 2: Transfer Learning using a Pre-trained CNN

## Aim
To implement Transfer Learning using a pre-trained CNN model (MobileNetV2/ResNet50) on the MNIST handwritten digit dataset.

## Objective
- Load the MNIST dataset.
- Display sample handwritten digit images.
- Convert grayscale images into RGB.
- Resize images to the required size.
- Load a pre-trained CNN model.
- Replace the final classification layer.
- Train the transfer learning model.
- Evaluate the model.
- Display predictions.

## Dataset
- **Dataset:** MNIST Handwritten Digits
- **Classes:** 10
- **Image Size:** 28×28 → 96×96 (or 224×224)
- **Model:** MobileNetV2 (Fast Version)

## Libraries Used
- TensorFlow
- NumPy
- Matplotlib
- Keras Applications

## Expected Outcome
The transfer learning model should accurately recognize handwritten digits and display the predicted and actual labels.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train[:5000]
y_train = y_train[:5000]

x_test = x_test[:1000]
y_test = y_test[:1000]

print("Training Images:", x_train.shape)
print("Testing Images :", x_test.shape)

# ----------------------------------------------------------
# Display Sample Images
# ----------------------------------------------------------

plt.figure(figsize=(12,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(y_train[i])
    plt.axis("off")

plt.show()

# ----------------------------------------------------------
# Convert Grayscale to RGB
# ----------------------------------------------------------

x_train = np.expand_dims(x_train,-1)
x_test = np.expand_dims(x_test,-1)

x_train = tf.image.grayscale_to_rgb(tf.convert_to_tensor(x_train))
x_test = tf.image.grayscale_to_rgb(tf.convert_to_tensor(x_test))

# ----------------------------------------------------------
# Resize Images
# ----------------------------------------------------------

IMG_SIZE = 96

x_train = tf.image.resize(x_train,(IMG_SIZE,IMG_SIZE))
x_test = tf.image.resize(x_test,(IMG_SIZE,IMG_SIZE))

x_train = preprocess_input(x_train)
x_test = preprocess_input(x_test)

# ----------------------------------------------------------
# Load Pretrained MobileNetV2
# ----------------------------------------------------------

base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(96,96,3)
)

base_model.trainable = False

# ----------------------------------------------------------
# Build Transfer Learning Model
# ----------------------------------------------------------

x = base_model.output
x = GlobalAveragePooling2D()(x)
output = Dense(10, activation="softmax")(x)

model = Model(base_model.input, output)

# ----------------------------------------------------------
# Compile
# ----------------------------------------------------------

model.compile(
    optimizer=Adam(0.0001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# ----------------------------------------------------------
# Train
# ----------------------------------------------------------

model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=2,
    batch_size=32
)

# ----------------------------------------------------------
# Evaluate
# ----------------------------------------------------------

loss, accuracy = model.evaluate(x_test, y_test)

print("\nTest Loss :", loss)
print("Test Accuracy :", accuracy)

# ----------------------------------------------------------
# Predictions
# ----------------------------------------------------------

predictions = model.predict(x_test[:10])
predicted_labels = np.argmax(predictions, axis=1)

# ----------------------------------------------------------
# Display Predictions
# ----------------------------------------------------------

plt.figure(figsize=(12,5))

for i in range(10):

    img = (x_test[i] + 1) / 2

    plt.subplot(2,5,i+1)
    plt.imshow(img)

    plt.title(
        f"P:{predicted_labels[i]}\nA:{y_test[i]}",
        fontsize=10
    )

    plt.axis("off")

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Model Summary
# ----------------------------------------------------------

model.summary()
```
