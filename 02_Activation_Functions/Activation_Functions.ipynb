# Comparison of Activation Functions

## 🎯 Objective

Visualize the most commonly used activation functions.

---

## 📚 Theory

Activation functions introduce non-linearity into neural networks.

Without activation functions, deep neural networks behave like linear models.

---

## 📌 Activation Functions

- Sigmoid
- Tanh
- ReLU

---

## 🧮 Formula

### Sigmoid

σ(x)=1/(1+e⁻ˣ)

### Tanh

tanh(x)

### ReLU

max(0,x)

## 💻 Implementation

## 📈 Graph

---

## 📖 Conclusion

Different activation functions behave differently and are used for different applications.

```python
import numpy as np
import matplotlib.pyplot as plt

#Input values
x=np.linspace(-10,10,200)

#sigmoid
sigmoid = 1/(1+np.exp(-x))

#Tanh
tanh = np.tanh(x)

#Relu
relu = np.maximum(0,x)

#plot graph
plt.figure(figsize=(8,6))
plt.plot(x,sigmoid,label='Sigmoid', linewidth=2)
plt.plot(x,tanh,label='Tanh', linewidth=2)
plt.plot(x,relu,label='Relu', linewidth=2)

plt.title("Comparison of Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("output")
plt.legend()
plt.grid(True)
plt.show()
```
