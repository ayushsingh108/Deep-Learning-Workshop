## 📚 Theory

A Convolutional Neural Network (CNN) is a deep learning model specially designed for image processing.

Instead of directly connecting every pixel to every neuron, CNNs extract important features such as edges, textures, and shapes using convolution filters.

CNNs are widely used in:

- Image Classification
- Face Recognition
- Medical Imaging
- Object Detection
- Self-driving Cars

---

---

## 📖 Layer Explanation

### 1. Convolution Layer

Extracts important image features using filters.

Example:

- Edges
- Corners
- Patterns
- Textures

---

### 2. ReLU Activation

Formula

ReLU(x) = max(0,x)

Purpose

Introduces non-linearity into the network.

---

### 3. Max Pooling

Reduces image size while preserving important information.

Example

2×2 Max Pooling

Input

1 4

3 2

Output = 4

---

### 4. Flatten Layer

Converts the feature maps into a one-dimensional vector before passing them to fully connected layers.

---

### 5. Dense Layer

Acts like a traditional Artificial Neural Network.

Performs final classification.

---

### 6. Softmax Layer

Produces probabilities for each class.

The class with the highest probability becomes the prediction.

## 📖 Conclusion

CNNs automatically learn image features without manual feature extraction, making them one of the most powerful models for computer vision tasks.

## 🔑 Key Takeaways

- Convolution extracts features.
- ReLU introduces non-linearity.
- Pooling reduces computation.
- Flatten converts images into vectors.
- Dense layers perform classification.
- Softmax produces class probabilities.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Build CNN Model
model = Sequential()

# Convolution Layer
model.add(Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(28,28,1)
))

# Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten Layer
model.add(Flatten())

# Hidden Dense Layer
model.add(Dense(128, activation='relu'))

# Output Layer
model.add(Dense(10, activation='softmax'))

# Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display Model Architecture
model.summary()
```
