## Program 1: Convert an Image to Grayscale

### Description

This program reads a color image and converts it into a grayscale image.

Grayscale images contain only intensity values instead of three color channels (RGB/BGR). They are commonly used because they reduce computational complexity and are sufficient for many computer vision tasks.

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("cat.jpeg")

if image is None:
    print("Error: Image not found.")
else:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    print("Original Image")
    cv2_imshow(image)

    print("Grayscale Image")
    cv2_imshow(gray)
```

## Program 2: Rotate an Image by 90°

### Description

OpenCV provides a simple function to rotate an image by fixed angles such as 90°, 180°, and 270°.

This method is useful when a fixed rotation is required.

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image
image = cv2.imread("cat.jpeg")

# Check if image is loaded
if image is None:
    print("Error: Image not found. Please check the file path.")
else:
    # Rotate image 90 degrees clockwise
    rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

    # Display images
    print("Original Image")
    cv2_imshow(image)

    print("Rotated Image (90° Clockwise)")
    cv2_imshow(rotated)
```

## Program 3: Rotate an Image using a Rotation Matrix

### Description

Instead of rotating by fixed angles, OpenCV allows rotation by any angle using an affine transformation.

The function `cv2.getRotationMatrix2D()` creates the rotation matrix, and `cv2.warpAffine()` applies the transformation to the image.

This technique is widely used in image augmentation for deep learning.

```python
import cv2
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
import time

# Read the image
image = cv2.imread("cat.jpeg")

if image is None:
    print("Error: Image not found. Please check the file path.")
else:
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Get image dimensions
    h, w = gray.shape[:2]
    center = (w // 2, h // 2)

    print("Original Image")
    cv2_imshow(image)

    print("Grayscale Image")
    cv2_imshow(gray)

    # Rotate from 0° to 360°
    for angle in range(0, 361, 5):
        matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
        rotated = cv2.warpAffine(gray, matrix, (w, h))

        clear_output(wait=True)
        print(f"Rotated Image: {angle}°")
        cv2_imshow(rotated)

        time.sleep(0.03)   # 30 ms delay
```
