## 💻 Implementation

---

## 📖 Conclusion

---

## 🔑 Key Takeaways

- cv2.rectangle()
- cv2.putText()

```python
import cv2
from google.colab.patches import cv2_imshow

# Read image
image = cv2.imread("image.jpg")

cv2.rectangle(image, (250, 120), (500, 550), (255,0,0), 3)

cv2.putText(image,
            "Cat",
            (260, 110),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (0,0,255),
            2)

# Display image
cv2_imshow(image)
```
