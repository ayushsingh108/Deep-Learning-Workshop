# Intersection over Union (IoU)

## 🎯 Objective

Calculate the Intersection over Union (IoU) between two bounding boxes.

---

## 📚 Theory

IoU measures the overlap between the predicted bounding box and the ground truth.

IoU = Intersection Area / Union Area

Higher IoU indicates better localization.

---

## 💻 Implementation

Compute IoU for one prediction.

---

## 📊 Expected Output

IoU score.

---

## 📖 Conclusion

IoU is the standard metric for evaluating object detection models.

---

## 🔑 Key Takeaways

- Perfect overlap = IoU 1
- No overlap = IoU 0

```python
def calculate_iou(box1, box2):
    x_left = max(box1[0], box2[0])
    y_top = max(box1[1], box2[1])
    x_right = min(box1[2], box2[2])
    y_bottom = min(box1[3], box2[3])

    intersection = max(0, x_right - x_left) * max(0, y_bottom - y_top)

    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union = area1 + area2 - intersection
    iou = intersection / union

    return iou


ground_truth = [100, 100, 200, 200]
prediction = [120, 120, 220, 220]

print("IoU =", round(calculate_iou(ground_truth, prediction), 2))
```
