# Precision Calculation

## 🎯 Objective

Calculate Precision for an object detection model.

---

## 📚 Theory

Precision measures how many predicted objects are actually correct.

Precision = TP / (TP + FP)

---

## 💻 Implementation

Compute Precision using TP and FP.

---

## 📊 Expected Output

Precision value.

---

## 📖 Conclusion

Higher Precision means fewer false detections.

---

## 🔑 Key Takeaways

- TP = True Positive
- FP = False Positive

```python
TP = 8
FP = 2
precision = TP/(TP+FP)
print("Precision =", precision)
```
