# Display Images in Google Colab

## 🎯 Objective

Learn different ways to display images in Google Colab using OpenCV.

---

## 📚 Theory

OpenCV uses `cv2.imread()` to load images from disk.

Since `cv2.imshow()` does not work in Google Colab, Google provides `cv2_imshow()` to display images.

This notebook demonstrates two methods:

1. Display a local image stored in the Colab runtime.
2. Download and display an image directly from the internet.

---

## 📦 Libraries Used

- OpenCV (`cv2`)
- os
- urllib.request
- google.colab.patches

---

## 💻 Implementation

### Method 1
Read and display a local image.

### Method 2
Download an image from the internet and display it.

---

## 📊 Expected Output

The selected image is displayed successfully in the notebook.

---

## 📖 Conclusion

Both methods load and display images using OpenCV.

- Use **Method 1** when working with your own uploaded images.
- Use **Method 2** when creating notebooks that should run without requiring users to upload files.

---

## 🔑 Key Takeaways

- `cv2.imread()` reads an image from disk.
- `cv2_imshow()` displays images in Google Colab.
- Always check whether the image loaded successfully before processing it.

##Method 1

```python
import cv2
import os
from google.colab.patches import cv2_imshow

print("Current Folder:", os.getcwd())

image = cv2.imread("images.jpg")

if image is None:
    print("ERROR: images.jpg not found!")
else:
    cv2_imshow(image)
```

##Method 2

```python
import cv2
import urllib.request
from google.colab.patches import cv2_imshow

image_url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
urllib.request.urlretrieve(image_url, "face.jpg")

image = cv2.imread("face.jpg")

cv2_imshow(image)
```
