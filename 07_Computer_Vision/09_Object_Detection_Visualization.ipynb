# Object Detection Visualization

## 🎯 Objective

Visualize object detection results by drawing bounding boxes and confidence scores.

---

## 📚 Theory

Object detection models return:

- Bounding Box
- Class Label
- Confidence Score

These predictions are drawn on the image for visualization.

---

## 📦 Import Libraries

- OpenCV
- NumPy

---

## 💻 Implementation

Create a blank image.

Draw a bounding box.

Display the object label and confidence score.

---

## 📊 Expected Output

Image showing a detected object with its confidence score.

---

## 📖 Conclusion

Visualization helps interpret the predictions of object detection models.

---

## 🔑 Key Takeaways

- Bounding boxes indicate object location.
- Labels identify the object.
- Confidence scores indicate prediction certainty.

## For Windows



```python
import cv2
import numpy as np

# Create a white image
image = np.ones((400, 600, 3), dtype=np.uint8) * 255

# Draw a green rectangle
cv2.rectangle(image, (60, 60), (180, 300), (0, 255, 0), 2)

# Add text
cv2.putText(
    image,
    "Person 98%",
    (50, 45),
    cv2.FONT_HERSHEY_SIMPLEX,
    0.7,
    (0, 0, 255),
    2
)

# Display the image
cv2.imshow("Object Detection", image)

# Wait until a key is pressed
cv2.waitKey(0)

# Close all OpenCV windows
cv2.destroyAllWindows()
```
