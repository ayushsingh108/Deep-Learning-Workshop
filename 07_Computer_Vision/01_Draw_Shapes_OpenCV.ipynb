# Drawing Shapes using OpenCV

## 🎯 Objective

Create a blank image and draw basic shapes and text using OpenCV.

---

## 📚 Theory

OpenCV provides functions for drawing geometric shapes such as rectangles, circles and lines.

It is widely used for image annotation and computer vision applications.

---

## 📦 Import Libraries

- OpenCV
- NumPy

---

## 💻 Implementation

- Create a blank image
- Draw a rectangle
- Add text
- Display image

---

## 📊 Expected Output

White image containing:

- Blue rectangle
- Red text

---

## 📖 Conclusion

OpenCV can create and annotate images programmatically.

---

## 🔑 Key Takeaways

- cv2.rectangle()
- cv2.putText()
- cv2.imshow()

## For Windows

```python
import cv2
import numpy as np

image =np.ones((400,500,3), dtype=np.uint8)*255

cv2.rectangle(image, (100,100), (300,250), (255,0,0), 3)

cv2.putText(image,"Cat",(110,70),
            cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,0,255),2)

cv2.imshow("Image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##For Google Colab

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = np.ones((400,500,3), dtype=np.uint8) * 255

cv2.rectangle(image, (100,100), (300,250), (255,0,0), 3)

cv2.putText(image,
            "Cat",
            (110,70),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (0,0,255),
            2)

cv2_imshow(image)
```
