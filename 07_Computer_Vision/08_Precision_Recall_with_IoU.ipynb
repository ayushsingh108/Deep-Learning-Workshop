# Precision and Recall using IoU

## 🎯 Objective

Evaluate multiple object detections using IoU, Precision and Recall.

---

## 📚 Theory

A prediction is considered correct when IoU exceeds a predefined threshold.

The evaluation computes:

- True Positives
- False Positives
- False Negatives
- Precision
- Recall

---

## 📦 Import Libraries

No external libraries required.

---

## 💻 Implementation

- Calculate IoU
- Match predictions
- Compute TP, FP, FN
- Calculate Precision
- Calculate Recall

---

## 📊 Expected Output

- IoU values
- Precision
- Recall

---

## 📖 Conclusion

IoU together with Precision and Recall provides a complete evaluation of object detection performance.

---

## 🔑 Key Takeaways

- IoU threshold determines correct detections.
- Precision measures correctness.
- Recall measures completeness.

```python
def calculate_iou(box1, box2):
    x_left = max(box1[0], box2[0])
    y_top = max(box1[1], box2[1])
    x_right = min(box1[2], box2[2])
    y_bottom = min(box1[3], box2[3])

    intersection = max(0, x_right - x_left) * max(0, y_bottom - y_top)
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union = area1 + area2 - intersection
    return intersection / union

ground_truth = [
    [100, 100, 200, 200],
    [250, 100, 350, 220]
]

predictions = [
    [110,110,205,205],
    [260,110,340,210],
    [20,20,80,80]
]

tp = 0
fp = 0
matched_gt = []

for pred in predictions:
    matched = False
    for i, gt in enumerate(ground_truth):
        iou = calculate_iou(pred, gt)
        if iou >= 0.5 and i not in matched_gt:
            tp += 1
            matched = True
            matched_gt.append(i)
            print("IoU =", round(iou, 2))
            break
    if not matched:
        fp += 1

fn = len(ground_truth) - tp

precision = tp / (tp + fp)
recall = tp / (tp + fn)

print("TP =", tp)
print("FP =", fp)
print("FN =", fn)
print("Precision =", round(precision, 2))
print("Recall =", round(recall, 2))
```
