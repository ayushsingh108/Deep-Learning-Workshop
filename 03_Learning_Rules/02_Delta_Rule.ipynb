# Delta Learning Rule

## 🎯 Objective

Understand how weights are updated using prediction error.

---

## 📚 Theory

The Delta Rule minimizes prediction error by adjusting the weights.

---

## 🧮 Formula

ΔW = η × Error × Input

## 📖 Conclusion

Weights gradually converge towards the optimal solution.

```python
learning_rate= 0.1
weight = 1.0
x = 2
target = 6
for epoch in range(10):
  prediction = weight * x
  error = target - prediction
  weight = weight + learning_rate*error * x
  print(f"Epoch{epoch+1}")
  print("prediction= ", prediction)
  print("Error= ", error)
  print("Updated Weight= ", round(weight,3))
  print("--------------------------------")
```
