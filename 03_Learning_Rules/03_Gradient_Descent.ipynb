## 📚 Theory

## 📖 Conclusion

Gradient Descent finds the optimal weight and bias values.

```python
import numpy as np
X=np.array([1,2,3,4,5],dtype=float)
y=np.array([2,4,6,8,10],dtype=float)
weight=0
bias=0
learning_rate=0.01
epochs=100
for epoch in range(epochs):
  predictions=weight*X+bias
  error=predictions-y
  dw=(2/len(X))*np.sum(error*X)
  db=(2/len(X))*np.sum(error)
  weight=weight-learning_rate*dw
  bias=bias-learning_rate*db
print("Final Weight: ",round(weight,3))
print("Final Bias: ",round(bias,3))
test=7
prediction=weight*test+bias
print("Prediction For X=7: ",round(prediction,2))
```
