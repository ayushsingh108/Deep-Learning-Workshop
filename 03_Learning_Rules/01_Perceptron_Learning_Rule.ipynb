# Perceptron Learning Rule

## 🎯 Objective

Train a perceptron by updating its weights based on prediction error.

---

## 📚 Theory

The perceptron learning rule updates weights whenever the prediction is incorrect.

---

## 🧮 Formula

Weight = Weight + Learning Rate × Error × Input

Bias = Bias + Learning Rate × Error

---

## 📦 Dataset

circuit_data.csv

---

## 💻 Implementation

---

## 📊 Training Output

---

## 📖 Conclusion

The perceptron gradually learns the correct decision boundary.

```python
import pandas as pd
import numpy as np

data=pd.read_csv("circuit_data.csv")
print("Dataset\n")
print(data)

# Separate Inputs and output
X = data.iloc[:,0:3].values
y = data.iloc[:, 3].values

# Initialize weights and bias
weights = np.zeros(3)
bias = 0

learning_rate = 0.1
epochs = 20

for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        # Weighted Sum
        z = np.dot(X[i], weights) + bias

        # Step Activation Function
        if z >= 0:
            prediction = 1
        else:
            prediction = 0

        # Calculate Error
        error = y[i] - prediction

        # Update Weights
        weights = weights + learning_rate * error * X[i]

        # Update Bias
        bias = bias + learning_rate * error

        print("Input:", X[i],
              "Target:", y[i],
              "Prediction:", prediction,
              "Error:", error)

print("\nTraining Completed")
print("Final Weights =", weights)
print("Final Bias =", bias)

# -----------------------------
# Step 5 : Test New Input
# -----------------------------
new_input = np.array([1,1,0])

z = np.dot(new_input, weights) + bias

if z >= 0:
    output = 1
else:
    output = 0

print("\nNew Input :", new_input)
print("Weighted Sum :", round(z,2))
print("Predicted Output :", output)
```
