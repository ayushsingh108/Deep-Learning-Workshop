# Perceptron Learning for AND Gate

## 🎯 Objective

Implement the Perceptron Learning Algorithm to train a perceptron for the AND logic gate and verify its predictions after training.

---

## 📚 Theory

A Perceptron is the simplest type of Artificial Neural Network used for binary classification.

The perceptron learns by adjusting its weights and bias whenever it makes an incorrect prediction. This process continues for multiple epochs until the model correctly classifies all training examples.

The AND Gate is a **linearly separable** problem, making it suitable for a single perceptron.

---

## 🧮 Perceptron Formula

### Weighted Sum

z = w₁x₁ + w₂x₂ + b

### Step Activation Function

Output = 1, if z ≥ 0

Output = 0, otherwise

### Weight Update Rule

w = w + η × Error × Input

### Bias Update Rule

b = b + η × Error

Where:

- **w** = Weight
- **η** = Learning Rate
- **Error** = Target − Prediction
- **b** = Bias

## 💻 Implementation

The program performs the following steps:

4. Apply the Step Activation Function.
5. Calculate prediction error.
6. Update weights and bias.
7. Repeat for multiple epochs.
8. Test the trained perceptron.

## 🔑 Key Takeaways

- The AND Gate is a linearly separable problem.
- A single perceptron can solve the AND Gate.
- The Perceptron Learning Rule updates weights using prediction error.
- Training continues until the model learns the correct decision boundary.
- This example demonstrates the fundamentals of supervised learning in neural networks.

```python
import numpy as np

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,0,0,1])   # AND Gate

weights = np.zeros(2)
bias = 0

learning_rate = 0.1
epochs = 10

def step(x):
    return 1 if x >= 0 else 0

for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        net = np.dot(X[i], weights) + bias
        output = step(net)

        error = y[i] - output

        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print("Input:", X[i],
              "Target:", y[i],
              "Output:", output,
              "Error:", error)

print("\nFinal Weights:", weights)
print("Final Bias:", bias)

print("\nTesting")

for sample in X:
    result = step(np.dot(sample, weights)+bias)
    print(sample, "->", result)
```
