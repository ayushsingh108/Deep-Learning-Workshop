# L2 Regularization

## 🎯 Objective

Reduce overfitting using L2 Regularization.

---

## 📚 Theory

L2 Regularization penalizes large weights by adding a penalty term to the loss function.

---

## 🧮 Formula

Loss = Original Loss + λW²

---

## 💻 Implementation

---

## 📊 Output

---

## 📖 Conclusion

L2 Regularization improves model generalization.

```python
import numpy as np
weight = 5.0
gradient = 8.0
learning_rate =0.1
lambda_reg = 0.2
print("Initial Weight: ", weight)

regularization =2*lambda_reg*weight
total_gradient = gradient + regularization
weight = weight - learning_rate*total_gradient
print("Regularization Term:",regularization)
print("Updated Weight: ", round(weight,2))
```
