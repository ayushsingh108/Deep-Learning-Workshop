## 📖 Conclusion

OpenCV can manipulate images efficiently with built-in functions.

```python
import cv2
image = cv2.imread("cat.jpg")
related = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow("original Image", image)
cv2.imshow("Rotated Image", related)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
