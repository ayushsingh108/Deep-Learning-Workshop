# Electronic Circuit ON/OFF using Perceptron

## 🎯 Objective

Simulate an electronic switch using a perceptron.

---

## 📚 Theory

A perceptron can behave like a digital switch.

Binary inputs are multiplied by weights and compared with a threshold.

---

## 🧮 Formula

z = w₁x₁ + w₂x₂ + b

---

## 📌 Inputs

- Switch 1
- Switch 2

---

## 📦 Import Libraries

---

## 💻 Implementation

---

## 📊 Output

---

## 📖 Conclusion

The perceptron correctly predicts whether the circuit is ON or OFF.

```python
import numpy as np;

#input values
x1= 1
x2 = 0

#weights
w1 = 0.8
w2 = 0.5

#bias
bias = -0.4

#weighted Sum
z = (w1 * x1) + (w2 * x2) + bias

#Step Action
if z>=0:
  output = 1
else:
  output = 0

print("Weighted Sum = ", round(z,2))
print("Decision =" , output)
```
