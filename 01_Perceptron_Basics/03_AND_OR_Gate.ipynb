## 🎯 Objective

Implement logical AND and OR gates using a perceptron.

## 📚 Theory

Logic gates are the building blocks of digital circuits.

A perceptron can implement linearly separable gates by choosing suitable weights and bias.

## 📖 Conclusion

A single perceptron can implement linearly separable logic gates.

```python
import numpy as np

def step_function(z):
  """Implements the step activation function."""
  if z >= 0:
    return 1
  else:
    return 0

def perceptron(x1, x2, w1, w2, bias):
  """A simple perceptron model."""
  inputs = np.array([x1, x2])
  weights = np.array([w1, w2])
  z = np.dot(inputs, weights) + bias
  return step_function(z)

print("--- AND Gate Perceptron ---")
# Weights and bias for an AND gate
# A common set is w1=0.5, w2=0.5, bias=-0.7
# Or w1=1, w2=1, bias=-1.5 (as used here for clarity)
and_w1, and_w2, and_bias = 1, 1, -1.5

print(f"Inputs (0, 0): Output = {perceptron(0, 0, and_w1, and_w2, and_bias)}")
print(f"Inputs (0, 1): Output = {perceptron(0, 1, and_w1, and_w2, and_bias)}")
print(f"Inputs (1, 0): Output = {perceptron(1, 0, and_w1, and_w2, and_bias)}")
print(f"Inputs (1, 1): Output = {perceptron(1, 1, and_w1, and_w2, and_bias)}")

print("\n--- OR Gate Perceptron ---")
# Weights and bias for an OR gate
# A common set is w1=0.5, w2=0.5, bias=-0.2
# Or w1=1, w2=1, bias=-0.5 (as used here for clarity)
or_w1, or_w2, or_bias = 1, 1, -0.5

print(f"Inputs (0, 0): Output = {perceptron(0, 0, or_w1, or_w2, or_bias)}")
print(f"Inputs (0, 1): Output = {perceptron(0, 1, or_w1, or_w2, or_bias)}")
print(f"Inputs (1, 0): Output = {perceptron(1, 0, or_w1, or_w2, or_bias)}")
print(f"Inputs (1, 1): Output = {perceptron(1, 1, or_w1, or_w2, or_bias)}")
```
