# Loan Decision System using Perceptron

## 🎯 Objective

Build a simple perceptron that predicts whether a loan
should be approved or rejected.

---

## 📚 Theory

A perceptron is the simplest neural network.

It calculates

z = Σ(w × x) + b

If z ≥ 0 → Output = 1

Else → Output = 0

---

## 🧮 Formula

z = w₁x₁ + w₂x₂ + w₃x₃ + b

Output = Step(z)

---

## 📦 Import Libraries

```python
import numpy as np;

#input values
income = 0.8
credit_score = 0.9
employment = 1.0

#weights
w1 = 0.4
w2 = 0.5
w3 = 0.3

#bias
bias = -0.6

#weighted Sum
z = (w1 * income) + (w2 * credit_score) + (w3 * employment) + bias

#Step Action
if z>=0:
  output = 1
else:
  output = 0

print("Weighted Sum = ", round(z,2))
print("Decision =" , output)
```
