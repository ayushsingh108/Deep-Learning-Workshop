# Feature Maps in CNN

## 🎯 Objective

Visualize feature maps generated by the first convolution layer.

---

## 📚 Theory

Each convolution filter extracts different image features.

The output of each filter is called a feature map.

Feature maps help us understand what a CNN is learning.

---

## 📦 Import Libraries

- TensorFlow
- NumPy
- Matplotlib

---

## 💻 Implementation

Build a CNN.

Extract feature maps.

Display them using Matplotlib.

---

## 📊 Expected Output

Original Image

Feature Maps

---

## 📖 Conclusion

Feature maps show how CNN filters detect different image characteristics.

---

## 🔑 Key Takeaways

- Different filters learn different patterns.
- Feature maps help interpret CNN behavior.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), _ = tf.keras.datasets.mnist.load_data()
x_train = x_train / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D((2,2))
])

feature_model1 = tf.keras.Model(
    inputs=model.inputs,
    outputs=model.layers[0].output
)
feature_maps = feature_model1.predict(x_train[:1])
plt.figure(figsize=(10,5))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(feature_maps[0,:,:,i], cmap='gray')
    plt.axis('off')
plt.show()
```
