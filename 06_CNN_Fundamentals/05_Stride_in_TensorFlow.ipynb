# Stride in TensorFlow

## 🎯 Objective

Compare CNN output shapes for different stride values.

---

## 📚 Theory

Changing stride changes the size of the feature maps.

---

## 📦 Import Libraries

- TensorFlow

---

## 💻 Implementation

Build two Conv2D layers.

Stride = 1

Stride = 2

Compare model summaries.

---

## 📊 Expected Output

Two model summaries with different output sizes.

---

## 📖 Conclusion

Stride directly affects output dimensions.

---

## 🔑 Key Takeaways

- Stride = 1 preserves more information.
- Stride = 2 reduces feature map size.

```python
import tensorflow as tf
model1 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(
        filters=1,
        kernel_size=(3,3),
        strides=1,
        input_shape=(28,28,1)
    )
])

print("Stride = 1")
model1.summary()
model2 = tf.keras.Sequential([
    tf.keras.layers.Conv2D(
        filters=1,
        kernel_size=(3,3),
        strides=2,
        input_shape=(28,28,1)
    )
])

print("\nStride = 2")
model2.summary()
```
